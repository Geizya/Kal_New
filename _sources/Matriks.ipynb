### Definisi Metriks
susunan atau sekumpulan elemen (biasanya angka atau variabel) yang disusun dalam baris dan kolom yang membentuk suatu tabel persegi panjang atau kotak. Dalam notasi matematika, matriks sering kali dituliskan dalam bentuk huruf kapital seperti
A,B dan M

### Konsep dasar matriks
Notasi matriks biasanya ditulis dalam huruf kapital tebal seperti A, B, C, dll, dan elemen-elemen dalam matriks ditulis dalam tanda kurung atau kurung siku

### Operasi Aritmatika Metriks
Operasi aritmatika matriks adalah proses perhitungan matematika yang melibatkan satu atau lebih matriks, di mana setiap elemen dalam matriks dihitung sesuai dengan aturan operasi matematika dasar, seperti penjumlahan, pengurangan, perkalian, dan pembagian (meskipun pembagian langsung jarang digunakan pada matriks).
#### Syarat aritmatika metriks
1.Penjumlahan dan Pengurangan: Ukuran matriks harus sama.
2.Perkalian Matriks Jumlah kolom matriks pertama harus sama dengan jumlah baris matriks kedua.

### 1. Penjumlahan Metriks
Matriks harus memiliki ukuran yang sama (jumlah baris dan jumlah kolom harus sama).

$$
a=
\begin{bmatrix}
a & b\\
c & d\\
\end{bmatrix}
\quad
$$

$$
b=
\begin{bmatrix}
e & f\\
g & h\\
\end{bmatrix}
\quad
$$

maka:
$$
A+B=
\begin{array}{cc}
(a\ +\  e\ ) & (b\ +\ f\ ) &
(c\ +\  g\ ) & (d\ +\ h\ )
\end{array}
$$

jawab:
$$
AB=
\begin{bmatrix}
3+1 & 5+4\\
7+6 & 2+3\\
\end{bmatrix}
\quad
$$
$$
A\ B\ =
\begin{bmatrix}
4 & 9\\
13 & 5\\
\end{bmatrix}
\quad
$$


### 2. Pengurangan Metriks
operasi aritmatika untuk mengurangkan dua matriks dengan ukuran yang sama. Pengurangan dilakukan dengan cara mengurangkan elemen-elemen yang posisinya sama di kedua matriks.

$$
a=
\begin{bmatrix}
a & b\\
c & d\\
\end{bmatrix}
\quad
$$

$$
b=
\begin{bmatrix}
e & f\\
g & h\\
\end{bmatrix}
\quad
$$

maka:
$$
A+B=
\begin{array}{cc}
(a\ -\  e\ ) & (b\ -\ f\ ) &
(c\ -\  g\ ) & (d\ -\ h\ )
\end{array}
$$

jawab:
$$
AB=
\begin{bmatrix}
7-4 & 3-8\\
2-1 & 5-6\\
\end{bmatrix}
\quad
$$
$$
A\ B\ =
\begin{bmatrix}
5 & 3\\
-2 & 2\\
\end{bmatrix}
\quad
$$

### 3. Perkalian Metriks
Perkalian matriks adalah operasi aritmatika pada dua matriks di mana elemen-elemen dari baris matriks pertama dikalikan dengan elemen-elemen dari kolom matriks kedua, lalu hasilnya dijumlahkan.

contoh perkalian 2 metriks atau lebih.
jawab:
$$

$$

$$

```python
##CODE PENJUMLAHAN METRIKS

A = [
    [3, 1],
    [5, 4]
]
B = [
    [7, 6],
    [2, 3]
]
result = [
    [0, 0],
    [0, 0]
]
for i in range(len(A)):
    for j in range(len(A[0])):
        result[i][j] = A[i][j] + B[i][j]
print("Hasil penjumlahan matriks A + B:")
for row in result:
    print(row)
```

```python
##CODE PENGURANGAN METRIKS

A = [
    [7, 4],
    [3, 8]
]

B = [
    [2, 3],
    [4, 5]
]

hasil = [
    [0, 0],
    [0, 0]
]

for i in range(2):        # 2 baris
    for j in range(2):    # 2 kolom
        hasil[i][j] = A[i][j] - B[i][j]

print("Hasil Pengurangan Matriks A - B:")
for row in hasil:
    print(row)
```

```python
##CODE PERKALIAN METRIKS
import numpy as np

A = np.array([[1, 0], [0, 1]])
B = np.array([[5, 6], [7, 8]])
C = np.array([[1, 1], [0, 1]])

# Perkalian A x B x C
result = A @ B @ C

print("Hasil Perkalian A x B x C:\n", result)
```

```python
## Perkalian 2 metriks atau lebih
def perkalian_matriks(A, B):
    # Ukuran: A(m x n), B(n x p)
    m = len(A)
    n = len(A[0])
    p = len(B[0])

    hasil = [[0 for _ in range(p)] for _ in range(m)]

    for i in range(m):
        for j in range(p):
            for k in range(n):
                hasil[i][j] += A[i][k] * B[k][j]

    return hasil

A = [
    [1, 0],
    [0, 1]
]

B = [
    [2, 3],
    [4, 5]
]

C = [
    [1, 1],
    [0, 1]
]

# Perkalian pertama: A x B
hasil1 = perkalian_matriks(A, B)

# Perkalian kedua: hasil1 x C
hasil2 = perkalian_matriks(hasil1, C)

print("Hasil A x B:")
for baris in hasil1:
    print(baris)

print("\nHasil (A x B) x C:")
for baris in hasil2:
    print(baris)
```

```python
## Perkalian skalar
M = [
    [1, 2],
    [3, 4]
]

# Skalar
k = 3

# Perkalian skalar ke setiap elemen matriks
hasil = [[k * elemen for elemen in baris] for baris in M]

# Tampilkan hasil
print("Hasil perkalian skalar:")
for row in hasil:
    print(row)
```
